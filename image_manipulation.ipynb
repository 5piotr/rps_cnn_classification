```python
from PIL import Image, ImageFilter 
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
no_images = 15
```

```python
plt.figure(figsize=(12,10))
grid_x = 6
grid_y = int(no_images/3)
for i in range(no_images):
    my_image_r = image.load_img('pictures\\' + str(i+1)+'.jpg',target_size=(60,90),color_mode="rgb")
    my_image = image.load_img('pictures\\' + str(i+1)+'.jpg',target_size=(60,90),color_mode="grayscale")
    my_image = my_image_r.filter(ImageFilter.FIND_EDGES) 
            
    plt.subplot(grid_y,grid_x,i*2+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)       
    plt.imshow(my_image_r)
    
    plt.subplot(grid_y,grid_x,i*2+2)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)       
    plt.imshow(my_image)
```
